```python


import numpy as np
import pyggdrasil as yg
```

```python
# make 4 chains of 1000 zeros
array = np.zeros((1000,4)).T +1
```

```python
rhats = yg.analyze.rhats(array)
```

```python
rhats
```

```python
# numpy bernourli function
ber_a =  np.random.binomial(1, 0.2, size=(4,1000)) *1.0
```

```python
rhats = yg.analyze.rhats(ber_a)
ess = yg.analyze.ess(ber_a)
```
